```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataframeFibo = pd.DataFrame({'Linguagem':['Python', 'Lua', 'Ruby'], 'Tempo(s)':[43.08, 23.36,43.35]})
dataframeStri = pd.DataFrame({'Linguagem':['Python', 'Lua', 'Ruby'], 'Tempo(s)':[0.04, 0.01, 0.08]})
dataframeBubb = pd.DataFrame({'Linguagem':['Python', 'Lua', 'Ruby'], 'Tempo(s)':[0.67,0.29,0.44]})
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12,4), squeeze=False)

dataframeFibo.plot.bar(x='Linguagem', y='Tempo(s)', title= "Fibonacci", ax=axes[0, 0])
dataframeStri.plot.bar(x='Linguagem', y='Tempo(s)', title= "SubString NÃºmero", ax=axes[0, 1])
dataframeBubb.plot.bar(x='Linguagem', y='Tempo(s)', title= "Bubble Sort", ax=axes[0, 2])
```
